```python
#1
class RealNumber:
    
    def __init__(self, r=0):
        self.__realValue = r  
    def getRealValue(self):
        return self.__realValue
    def setRealValue(self, r):
        self.__realValue = r
    def __str__(self):
        return 'RealPart: '+str(self.getRealValue()) 
class ComplexNumber(RealNumber):
  def __init__(self,r=1.0,i=1.0):
    super().__init__(float(r))
    self.imagine=float(i)
  def __str__(self):
    return f"{super().__str__()}\nImaginaryPart: {self.imagine}"
    
cn1 = ComplexNumber()
print(cn1)
print('---------')
cn2 = ComplexNumber(5,7)
print(cn2)

```

```python
#2
class RealNumber:
    def __init__(self, number=0):
        self.number = number
    def __add__(self, anotherRealNumber):
        return self.number + anotherRealNumber.number
    def __sub__(self, anotherRealNumber):
        return self.number - anotherRealNumber.number
    def __str__(self):
        return str(self.number)
class ComplexNumber(RealNumber):
  def __init__(self,r,i):
    if type(r)==int:
      super().__init__(r)
    else:
      super().__init__(r.number)
    self.i=i
  def __add__(self,imag):
    return ComplexNumber(super().__add__(imag),self.i+imag.i)
  def __sub__(self,imag):
    return ComplexNumber(super().__sub__(imag),self.i-imag.i)
  def __str__(self):
    return f"{super().__str__()} + {self.i}i"
r1 = RealNumber(3)
r2 = RealNumber(5)
print(r1+r2)
cn1 = ComplexNumber(2, 1)
print(cn1)
cn2 = ComplexNumber(r1, 5)
print(cn2)
cn3 = cn1 + cn2
print(cn3)
cn4 = cn1 - cn2
print(cn4)
```

```python
#3
class Account: 
    def __init__(self, balance):
        self._balance = balance
        
    def getBalance(self):
        return self._balance
class CheckingAccount(Account):
  numberOfAccount=0
  def __init__(self,balance=None):
    super().__init__(balance)
    CheckingAccount.numberOfAccount+=1
  def __str__(self):
     if self._balance==None:
       return f"Account Balance: 0.0"
     else:
       return f"Account Balance: {float(self.getBalance())}"
print('Number of Checking Accounts: ', CheckingAccount.numberOfAccount)
print(CheckingAccount())
print(CheckingAccount(100.00))
print(CheckingAccount(200.00))
print('Number of Checking Accounts: ', CheckingAccount.numberOfAccount)
```

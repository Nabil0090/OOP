```python
#5
var = input()
lower, upper, digit, special = (0,0,0,0)

for char in var:
  if char>='a' and char<='z':
    lower+=1
  elif char>='A' and char<='Z':
    upper+=1
  elif char>='0' and char<='9':
    digit+=1
  elif char in "_$#@":
    special+=1
if (lower>0 and upper>0 and digit>0 and special>0):
  print("OK")
else:
  empty = ""
  if (lower==0):
    empty +="Lowercase character missing, " 
  if (upper==0):
    empty +="Uppercase character missing, "
  if (digit==0):
    empty +="Digit missing, "
  if (special==0):
    empty +="Special character missing, "
  print(empty[:-2])



```

```python
#Fn1
def BMI_calculator(height, weight):
  #height_in_meters = height / 100
  BMI = weight / (height/100) ** 2
  if BMI < 18.5:
    return f"Score is {round(BMI,1)}. You are Underweight"
  elif 18.5 <= BMI <= 24.9:
    return f"Score is {round(BMI,1)}. You are Normal"
  elif 25 <= BMI <= 30:
    return f"Score is {round(BMI,1)}. You are Overweight"
  elif BMI > 30:
    return f"Score is {round(BMI,1)}. You are Obese"     
print(BMI_calculator(175,96))
```

```python
#fn2
def calculation_of_num(min, max, div):
  sum = 0
  for i in range(min, max):
    if i % div == 0:
      sum += i
  return sum    
print(calculation_of_num(0, 10, 2))
```

```python
#fn4
def name_check(check, new_dom, old_dom="kaaj.com"):
    old=check
    check=check.split("@")
    if check[1] != "new_domain":
        check[1] = new_dom

    new = check[0] + "@" + check[1]
    if new != old:
        print("Changed: ",new)
    else:
        print("Unchanged: ",new)

name_check("alice@kaaj.com", "sheba.xyz", "kaaj.com")
```

```python
#fn5
def palindrome_eval(var):
    for i in range(0, int(len(var)/2)):
        if var[i] != var[len(var)-i-1]:
          return 'Not a palindrome'  
    return 'Palindrome'   
print(palindrome_eval('hello'))   
```

```python
#1
class Calculator:
    def __init__(self):
        print("Let's Calculate")
    def add(self, value1, value2):
        return value1 + value2
    def subtract(self, value1, value2):
        return value1 - value2
    def multiply(self, value1, value2):
        return value1 * value2
    def divide(self, value1, value2):
        return value1 / value2

basic_calculation = Calculator()
value1 = int(input())
operator = input()
value2 = int(input())

if operator == "+":
    print(basic_calculation.add(value1, value2))
elif operator == "-":
    print(basic_calculation.subtract(value1, value2))
elif operator == "*":
    print(basic_calculation.multiply(value1, value2))
elif operator == "/":
    print(basic_calculation.divide(value1, value2))


```

```python
#5
class Shape:
  def __init__(self,shape_name,num1,num2):
      self.shape_name=shape_name
      self.side1=num1
      self.side2=num2
  def area(self):
    if self.shape_name=="Triangle":
      area=(self.side1*self.side2)/2
      print("Area:",area)
    elif self.shape_name=="Square":
      area=self.side1*self.side2
      print("Area:",area)
    elif self.shape_name=="Rhombus":
      area=(self.side1*self.side2)/2
      print("Area:",area)
    elif self.shape_name=="Rectangle":
      area=self.side1*self.side2
      print("Area:",area)
    else:
      area="Shape unknown"
      print("Area:",area)

triangle = Shape("Triangle",10,25)
triangle.area()
print("==========================")
square = Shape("Square",10,10)
square.area()
print("==========================")
rhombus = Shape("Rhombus",18,25)
rhombus.area()
print("==========================")
rectangle = Shape("Rectangle",15,30)
rectangle.area()
print("==========================")
trapezium = Shape("Trapezium",15,30)
trapezium.area()
```

```python
#6
class Calculator:
  def __init__(self):
    self.num1=0
    self.num2=0
    self.sign=""
  def calculate(self,num1,num2,sign):
    self.sign=sign
    self.num1=num1
    self.num2=num2
    if sign=="+":
      self.total=num1+num2
      return self.total
    elif sign=="-":
      self.total=num1-num2
      return self.total
    elif sign=="*":
      self.total=num1*num2
      return self.total
    elif sign=="/":
      self.total=num1/num2
      return self.total
  def showCalculation(self):
    if self.sign=="+":
      print(f"{self.num1}+{self.num2}={self.total}")
    elif self.sign=="-":
      print(f"{self.num1}-{self.num2}={self.total}")
    elif self.sign=="/":
      print(f"{self.num1}/{self.num2}={self.total}")
    elif self.sign=="*":
      print(f"{self.num1}*{self.num2}={self.total}")
c1 = Calculator()
print("==================")
val = c1.calculate(10, 20, '+')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 10, '-')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 5, '*')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 16, '/')
print("Returned value:", val)
c1.showCalculation()




```

```python
#7
class Student:
  def __init__(self,name,id,dept,gpa):
    self.name=name
    self.id=id
    self.dept=dept
    self.gpa=gpa
  def calculate_CGPA(self):
    final=0
    total=0
    for i in range(len(self.gpa)):
      final+=self.gpa[i]
      total+=1
    self.result= round(final/total,2)
  def print_details(self):
    print(f"Name:{self.name},ID:{self.id}\nDepartment:{self.dept}\nCGPA:{self.result}")
    if self.result>3.80:
      print("Your academic Standing is 'Highest Distinction'")
    elif self.result>3.65:
      print("Your academic Standing is 'High Distinction'")
    elif self.result>3.50:
      print(" Your academic Standing is 'Distinction'")
    elif self.result>2.00:
      print("Your academic Standing is Satisfactory")
    else:
      print("Sorry you cant graduate")
s1 = Student('Dora', '15995599','CSE', [4,3.7,3.7,4])
s1.calculate_CGPA()
print("==========================")
s1.print_details()
print("==========================")
s2 = Student('Pingu', '12312322', 'EEE', [1.7,1.3,1.3,1.3,1])
s2.calculate_CGPA()
print("==========================")
s2.print_details()
print("==========================") 
s3 = Student('Bob', '13311331', 'CSE', [2,3,3,3.7,2.7,2.7])
s3.calculate_CGPA()
print("==========================")
s3.print_details()


```

```python
#8
class Shinobi:
  def __init__(self,name,rank):
    self.name=name
    self.rank=rank
    self.salary=0
    self.mission=0
  def calSalary(self,count):
    self.mission=count
    if self.rank=="Genin":
      self.salary=self.mission*50
    elif self.rank=="Chunin":
      self.salary=self.mission*100
    else:
      self.salary=self.mission*500
  def changeRank(self,rank):
    self.rank=rank
  def printInfo(self):
    print(f"Name:{self.name}\nRank:{self.rank}\nNumber of Mission:{self.mission}\nSalary:{self.salary}")
naruto = Shinobi("Naruto", "Genin")
naruto.calSalary(5)
naruto.printInfo()
print('====================')
shikamaru = Shinobi('Shikamaru', "Genin")
shikamaru.printInfo()
shikamaru.changeRank("Chunin")
shikamaru.calSalary(10)
shikamaru.printInfo()
print('====================')
neiji = Shinobi("Neiji", "Jonin")
neiji.calSalary(5)
neiji.printInfo()
```

```python
#10

class UberEats:
    def __init__(self, name, number, location):
        self.name = name
        self.number = number
        self.location = location
        print(f"{self.name}, welcome to UberEats!")

    def add_items(self, item1, item2, price1, price2):
        self.menu = {item1: price1, item2: price2}
        self.total_price = price1 + price2

    def print_order_detail(self):
        return f"User details: Name: {self.name}, Phone: {self.number}, Address: {self.location}\nOrder: {self.menu}\nTotal Paid Amount: {self.total_price}"

order1 = UberEats("Shakib", "01719658xxx", "Mohakhali")
print("=========================")
order1.add_items("Burger", "Coca Cola", 220, 50)
print("=========================")
print(order1.print_order_detail())
print("=========================")
order2 = UberEats("Siam", "01719659xxx", "Uttara")
print("=========================")
order2.add_items("Pineapple", "Dairy Milk", 80, 70)
print("=========================")
print(order2.print_order_detail())
```
